```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("CloudETL").getOrCreate()
```

```python
# Read in data from S3 Buckets
import pandas
from pyspark import SparkFiles
url="https://atlantic-hurricanes.s3.amazonaws.com/Resources/atlantic.csv" #Enter you S3 bucket URL (can be found in the summary of your bucket item in AWS)
spark.sparkContext.addFile(url)
df = spark.read.csv(SparkFiles.get("atlantic.csv"), sep=",", header=True, inferSchema=True)



# Show DataFrame
df.head()
```
